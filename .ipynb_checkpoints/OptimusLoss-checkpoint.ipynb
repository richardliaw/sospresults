```python
import numpy as np

def loss(k, b):
    b0, b1, b2 = b
    return (b0 * k/100 + 0.1 * b1 + 0.5) **(-1) + b2 * 0.01
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use(["seaborn-white", "seaborn-poster"])
%matplotlib inline
for i in range(100):
    b0 = np.random.exponential(0.1)
    b1, b2 = np.random.rand(2)
    b = [b0, b1, b2]
    plt.plot([loss(k, b) for k in range(500)], label=b[0], alpha=0.8)
plt.grid(True)
plt.savefig("optimus-loss.pdf")
```

```python
plt.hist([loss(500, [np.random.exponential(0.1), np.random.rand(), np.random.rand() ]) for k in range(500)], bins=30, rwidth=0.9, alpha=0.8)
plt.grid(True)
plt.savefig("optimus-loss-distribution.pdf")
```
